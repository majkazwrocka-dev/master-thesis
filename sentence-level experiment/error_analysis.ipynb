## Comparison of errors on Sherlock

```python

bert = pd.read_csv("2all_folds_sentence_predictions.tsv", sep="\t")
turing = pd.read_csv("3Turing_all_folds_sentence_predictions.tsv", sep="\t")

bert = bert.rename(columns={"prediction": "bert_pred"})
turing = turing.rename(columns={"prediction": "turing_pred"})

df = pd.merge(bert, turing, on=["sentence", "gold_label"])

df_pos = df[df["gold_label"] == "POSITIVE"].copy()

df_pos["bert_correct"] = df_pos["bert_pred"] == df_pos["gold_label"]
df_pos["turing_correct"] = df_pos["turing_pred"] == df_pos["gold_label"]

shared_pos = df_pos[(~df_pos["bert_correct"]) & (~df_pos["turing_correct"])]

bert_only_pos = df_pos[(~df_pos["bert_correct"]) & (df_pos["turing_correct"])]
turing_only_pos = df_pos[(df_pos["bert_correct"]) & (~df_pos["turing_correct"])]

print("Shared POSITIVE mistakes:", len(shared_pos))
print("Unique POSITIVE mistakes (BERT only):", len(bert_only_pos))
print("Unique POSITIVE mistakes (TuringBERT only):", len(turing_only_pos))

print("\n--- Unique POSITIVE Mistakes (BERT only) ---")
for _, row in bert_only_pos.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()

print("\n--- Unique POSITIVE Mistakes (TuringBERT only) ---")
for _, row in turing_only_pos.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()
```

```python
df_neu = df[df["gold_label"] == "NEUTRAL"].copy()

df_neu["bert_correct"] = df_neu["bert_pred"] == df_neu["gold_label"]
df_neu["turing_correct"] = df_neu["turing_pred"] == df_neu["gold_label"]

shared_neu = df_neu[(~df_neu["bert_correct"]) & (~df_neu["turing_correct"])]

bert_only_neu = df_neu[(~df_neu["bert_correct"]) & (df_neu["turing_correct"])]
turing_only_neu = df_neu[(df_neu["bert_correct"]) & (~df_neu["turing_correct"])]

print("Shared NEUTRAL mistakes:", len(shared_neu))
print("Unique NEUTRAL mistakes (BERT only):", len(bert_only_neu))
print("Unique NEUTRAL mistakes (TuringBERT only):", len(turing_only_neu))

print("\n--- Unique NEUTRAL Mistakes (BERT only) ---")
for _, row in bert_only_neu.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()

print("\n--- Unique NEUTRAL Mistakes (TuringBERT only) ---")
for _, row in turing_only_neu.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()
```

```python
df_neg = df[df["gold_label"] == "NEGATIVE"].copy()

df_neg["bert_correct"] = df_neg["bert_pred"] == df_neg["gold_label"]
df_neg["turing_correct"] = df_neg["turing_pred"] == df_neg["gold_label"]

shared_neg = df_neg[(~df_neg["bert_correct"]) & (~df_neg["turing_correct"])]

bert_only_neg = df_neg[(~df_neg["bert_correct"]) & (df_neg["turing_correct"])]
turing_only_neg = df_neg[(df_neg["bert_correct"]) & (~df_neg["turing_correct"])]

print("Shared NEGATIVE mistakes:", len(shared_neg))
print("Unique NEGATIVE mistakes (BERT only):", len(bert_only_neg))
print("Unique NEGATIVE mistakes (TuringBERT only):", len(turing_only_neg))

print("\n--- Unique NEGATIVE Mistakes (BERT only) ---")
for _, row in bert_only_neg.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()

print("\n--- Unique NEGATIVE Mistakes (TuringBERT only) ---")
for _, row in turing_only_neg.iterrows():
    print(f"Sentence: {row['sentence']}")
    print(f"Gold: {row['gold_label']} | BERT: {row['bert_pred']} | TuringBERT: {row['turing_pred']}")
    print()
```
